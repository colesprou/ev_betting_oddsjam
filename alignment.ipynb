```python
def extract_team_name(bet_name):
    """
    Extracts the team name from a Bet Name by splitting on '+' or '-'.
    Handles NaN values gracefully.
    """
    if isinstance(bet_name, str):  # Ensure the input is a string
        return re.split(r'[+-]', bet_name, 1)[0].strip()
    return None  # Return None for NaN or non-string inputs
```

```python
def squeeze_spreads_with_team_check(spread_df):
    """
    Pair positive and negative spreads for opposing teams based on absolute line values,
    and drop rows where the team in Bet Name_positive matches Bet Name_negative.
    """
    # Step 1: Standardize `line` values to floats
    spread_df["line"] = pd.to_numeric(spread_df["line"], errors="coerce")

    # Step 2: Separate Positive and Negative spreads
    positive_spreads = spread_df[
        spread_df["Bet Name"].str.contains("\+", na=False)
    ][["Game ID", "Market Name", "line", "Bet Name", "Sportsbook", "Odds"]].rename(
        columns={"Bet Name": "Bet Name_A", "Odds": "Odds_A"}
    )

    negative_spreads = spread_df[
        spread_df["Bet Name"].str.contains("-", na=False)
    ][["Game ID", "Market Name", "line", "Bet Name", "Sportsbook", "Odds"]].rename(
        columns={"Bet Name": "Bet Name_B", "Odds": "Odds_B"}
    )

    # Step 3: Create a column for absolute line values
    positive_spreads["abs_line"] = positive_spreads["line"].abs()
    negative_spreads["abs_line"] = negative_spreads["line"].abs()

    # Step 4: Merge Positive and Negative spreads on absolute line values
    paired_spreads = pd.merge(
        positive_spreads,
        negative_spreads,
        on=["Game ID", "Market Name", "abs_line", "Sportsbook"],
        suffixes=("_A", "_B"),
        how="outer"
    )

    # Step 5: Extract team names from Bet Names
    paired_spreads["Team_positive"] = paired_spreads["Bet Name_A"].apply(extract_team_name)
    paired_spreads["Team_negative"] = paired_spreads["Bet Name_B"].apply(extract_team_name)

    # Step 6: Drop rows where the teams are the same
    paired_spreads = paired_spreads[
        paired_spreads["Team_positive"] != paired_spreads["Team_negative"]
    ]

    # Step 7: Drop auxiliary columns and sort the results
    paired_spreads = paired_spreads.drop(columns=["abs_line", "Team_positive", "Team_negative"]).sort_values(by=["Game ID", "Market Name"])

    return paired_spreads
```

```python
def process_player_props(player_props_df):
    """
    Process the player props dataframe to pair Over and Under bets.
    """
    # Separate Over and Under bets
    over_bets = player_props_df[
        ["Game ID", "Game Name","Market Name", "line", "Bet Name", "Player Name", "Sportsbook", "Odds"]
    ][player_props_df["Bet Name"].str.contains("Over", na=False)]

    under_bets = player_props_df[
        ["Game ID","Game Name", "Market Name", "line", "Bet Name", "Player Name", "Sportsbook", "Odds"]
    ][player_props_df["Bet Name"].str.contains("Under", na=False)]

    # Pair Over and Under bets by Game ID, Market Name, line, and Player Name
    paired_player_props = pd.merge(
        over_bets,
        under_bets,
        on=["Game ID","Game Name", "Market Name", "line", "Player Name", "Sportsbook"],
        suffixes=("_A", "_B")
    )
    
    return paired_player_props

# Apply the function to process the player props
paired_player_props = process_player_props(pdf)
```

```python
def process_totals(totals_df):
    """
    Process the totals dataframe to pair Over and Under bets.
    """
    # Separate Over and Under bets
    over_bets = totals_df[
        ["Game ID","Game Name", "Market Name", "line", "Bet Name", "Sportsbook", "Odds"]
    ][totals_df["Bet Name"].str.contains("Over", na=False)]

    under_bets = totals_df[
        ["Game ID", "Game Name","Market Name", "line", "Bet Name", "Sportsbook", "Odds"]
    ][totals_df["Bet Name"].str.contains("Under", na=False)]

    # Pair Over and Under bets by Game ID, Market Name, and line
    paired_totals = pd.merge(
        over_bets,
        under_bets,
        on=["Game ID", "Game Name","Market Name", "line", "Sportsbook"],
        suffixes=("_A", "_B")
    )
    
    return paired_totals


def combine_processed_data(paired_totals, paired_spreads,props):
    """
    Combine processed totals and spreads into one DataFrame.
    """
    # Combine the totals and spreads data
    combined_df = pd.concat([paired_totals, paired_spreads,props], ignore_index=True, sort=False)
    return combined_df
processed_totals = process_totals(totals_df)
processed_props=process_totals(pdf)
squeezed_spreads_with_teams = squeeze_spreads_with_team_check(spread_df)

final_combined_df = combine_processed_data(processed_totals, squeezed_spreads_with_teams,paired_player_props)
```

```python
def implied_probability(odds):
    """
    Convert American odds to implied probability.
    """
    if odds > 0:
        return 100 / (odds + 100)
    else:
        return abs(odds) / (abs(odds) + 100)

def calculate_no_vig(row):
    """
    Calculate no-vig probabilities for each row.
    """
    # Calculate implied probabilities for Pinnacle Odds
    prob_a = implied_probability(row[('Odds_A', f'{sharpBook}')])
    prob_b = implied_probability(row[('Odds_B', f'{sharpBook}')])
    
    # Calculate total implied probability
    total_prob = prob_a + prob_b
    
    # Normalize to remove vig
    no_vig_prob_a = prob_a / total_prob
    no_vig_prob_b = prob_b / total_prob
    
    return pd.Series([no_vig_prob_a, no_vig_prob_b])

# Apply to DataFrame
pivot_table_df[['No_Vig_Prob_A', 'No_Vig_Prob_B']] = pivot_table_df.apply(calculate_no_vig, axis=1)
```

```python
def calculate_fliff_implied_probability(odds):
    """
    Convert Fliff odds into implied probability.
    """
    if odds > 0:  # Positive odds
        return 100 / (odds + 100)
    else:  # Negative odds
        return abs(odds) / (abs(odds) + 100)

# Apply to Odds_A and Odds_B
pivot_table_df[f'{Sportsbook}_Implied_Prob_A'] = pivot_table_df[('Odds_A', f'{Sportsbook}')].apply(calculate_fliff_implied_probability)
pivot_table_df[f'{Sportsbook}_Implied_Prob_B'] = pivot_table_df[('Odds_B', f'{Sportsbook}')].apply(calculate_fliff_implied_probability)
```

```python
# Check if Bet A is +EV
pivot_table_df['Is_Positive_EV_A'] = pivot_table_df['No_Vig_Prob_A'] > pivot_table_df[f'{Sportsbook}_Implied_Prob_A']

# Check if Bet B is +EV
pivot_table_df['Is_Positive_EV_B'] = pivot_table_df['No_Vig_Prob_B'] > pivot_table_df[f'{Sportsbook}_Implied_Prob_B']
```

```python
sportsbook_column_A = f'Odds_A_{Sportsbook}'
sportsbook_column_B = f'Odds_B_{Sportsbook}'
sharpBook_column_A = f'Odds_A_{sharpBook}'
sharpBook_column_B = f'Odds_B_{sharpBook}'
checkBook_column_A = f'Odds_A_BetOnline'
checkBook_column_B = f'Odds_B_BetOnline'
checkBook2_column_A = f'Odds_A_DraftKings'
checkBook2_column_B = f'Odds_B_DraftKings'
checkBook3_column_A = f'Odds_A_Bookmaker'
checkBook3_column_B = f'Odds_B_Bookmaker'
checkBook4_column_A = f'Odds_A_Pinnacle'
checkBook4_column_B = f'Odds_B_Pinnacle'
checkBook5_column_A = f'Odds_A_Betcris'
checkBook5_column_B = f'Odds_B_Betcris'
checkBook6_column_A = f'Odds_A_Circa Vegas'
checkBook6_column_B = f'Odds_B_Circa Vegas'
checkBook7_column_A = f'Odds_A_Blue Book'
checkBook7_column_B = f'Odds_B_Blue Book'









def generate_simple_bet_recommendations(df):
    """
    Generate simple recommendations for positive EV bets, sorted by EV in descending order.
    """
    # Add EV columns to the DataFrame
    df['EV_A'] = (df['No_Vig_Prob_A_'] - df[f'{Sportsbook}_Implied_Prob_A_']) * 100
    df['EV_B'] = (df['No_Vig_Prob_B_'] - df[f'{Sportsbook}_Implied_Prob_B_']) * 100

    # Create a long-form DataFrame for sorting by EV
    recommendations_df = []
    for _, row in df.iterrows():
        if row['EV_A'] > 0:
            recommendations_df.append({
                'Game Name': row['Game Name_'],
                'Bet Name': row['Bet Name_A_'],
                'Market Name': row['Market Name_'],
                f'Sportsbook Odds': row[sportsbook_column_A],
                f'sharpBook Odds': row[sharpBook_column_A],
                #'BetOnline Odds': row[checkBook_column_A],
                'DraftKings Odds': row[checkBook2_column_A],
               'Bookmaker Odds': row[checkBook3_column_A],
                'Pinnacle Odds': row[checkBook4_column_A],
                #'Betcris Odds': row[checkBook5_column_A],
                'Circa Vegas Odds': row[checkBook6_column_A],
               'FanDuel Odds':row[checkBook7_column_A],

                'EV': row['EV_A']
            })
        if row['EV_B'] > 0:
            recommendations_df.append({
                'Game Name': row['Game Name_'],
                'Bet Name': row['Bet Name_B_'],
                'Market Name': row['Market Name_'],
                f'Sportsbook Odds': row[sportsbook_column_B],
                f'sharpBook Odds': row[sharpBook_column_B],
               # 'BetOnline Odds': row[checkBook_column_B],
                'DraftKings Odds': row[checkBook2_column_B],
               'Bookmaker Odds': row[checkBook3_column_B],
                'Pinnacle Odds': row[checkBook4_column_B],
                #'Betcris Odds': row[checkBook5_column_B],
                'Circa Vegas Odds': row[checkBook6_column_B],
                'FanDuel Odds':row[checkBook7_column_B],


                'EV': row['EV_B']
            })

    # Convert recommendations to DataFrame for sorting
    recommendations_df = pd.DataFrame(recommendations_df)
    recommendations_df.sort_values(by='EV', ascending=False, inplace=True)

    # Generate text recommendations
    recommendations = []
    for _, rec in recommendations_df.iterrows():
        recommendations.append(
            f"{rec['Game Name']}\n"
            f"{rec['Bet Name']}\n"
            f"{rec['Market Name']}\n"
            f"{Sportsbook} Odds: {rec['Sportsbook Odds']}\n"
            f"{sharpBook} Odds: {rec['sharpBook Odds']}\n"
            #f"BetOnline Odds: {rec['BetOnline Odds']}\n"
            f"DraftKings Odds: {rec['DraftKings Odds']}\n"
            f"BookMaker Odds: {rec['Bookmaker Odds']}\n"
            f"Pinnacle Odds: {rec['Pinnacle Odds']}\n"
            #f"Betcris Odds: {rec['Betcris Odds']}\n"
            f"Circa Vegas Odds: {rec['Circa Vegas Odds']}\n"
           f"FanDuel Odds: {rec['FanDuel Odds']}\n"

            f"EV: {rec['EV']:.2f}%\n"
        )

    return "\n".join(recommendations)  # Join recommendations with actual newlines

# Generate sorted recommendations
sorted_recommendations_text = generate_simple_bet_recommendations(positive_ev_bets)

# Print the sorted recommendations
print(sorted_recommendations_text)
```

```python
# Create a table structure similar to the one shown in the image
def create_comparison_table(df, user_sportsbook, sharp_sportsbooks, odds_threshold=2):
    # Filter data for the user sportsbook
    user_data = df[df['Sportsbook'] == user_sportsbook][
        ['Game ID', 'Game Name', 'Market Name', 'line', 'Odds_A','Player Name']
    ]
    user_data.rename(columns={'line': 'Line User Sportsbook', 'Odds_A': 'Price User Sportsbook'}, inplace=True)

    # Initialize the comparison table with user sportsbook data
    comparison_table = user_data.copy()

    # Iterate over sharp sportsbooks and merge data
    for i, sharp_sportsbook in enumerate(sharp_sportsbooks, start=1):
        sharp_data = df[df['Sportsbook'] == sharp_sportsbook][
            ['Game ID', 'Market Name', 'line', 'Odds_A', "Player Name"]
        ]
        sharp_data.rename(
            columns={
                'line': f'Line Sharp Sportsbook {i}',
                'Odds_A': f'Price Sharpbook {i}'
            },
            inplace=True
        )
        # Merge sharp sportsbook data into the comparison table
        comparison_table = comparison_table.merge(
            sharp_data,
            on=['Game ID', 'Market Name','Player Name'],
            how='left'
        )

    # Filter rows where odds are similar between user sportsbook and sharp sportsbooks
    for i in range(1, len(sharp_sportsbooks) + 1):
        comparison_table = comparison_table[
            abs(comparison_table['Price User Sportsbook'] - comparison_table[f'Price Sharpbook {i}']) <= odds_threshold
        ]

    return comparison_table

# Example: Define user sportsbook and sharp sportsbooks
user_sportsbook = "MyBookie"
sharp_sportsbooks = ["DraftKings","Circa Vegas","Pinnacle" ]

# Generate the comparison table
comparison_table = create_comparison_table(final_combined_df, user_sportsbook, sharp_sportsbooks,odds_threshold=5)
```
